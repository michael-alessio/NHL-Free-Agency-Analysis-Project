```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
import numpy as np

pagination_page_list = []

for i in range(1,13):
    pagination_page_link = "https://www.spotrac.com/nhl/free-agents/_/year/" + str(i+2011) + "/status/signed/sort/contract_value"
    pagination_page_list.append(pagination_page_link)

data = []
year_count=2012

for current_page_link in pagination_page_list:

    print("*****NOW SCRAPING PAGE:",current_page_link)
    current_page_resp = requests.get(current_page_link)
    current_page_soup = BeautifulSoup(current_page_resp.content, "html.parser")

    row_tags = current_page_soup.find_all('tr')
    
    
    for row in row_tags:
        year=year_count
        try:
            name = row.find('a').get_text().strip()
        except:
            name = "NA"
        try:
            position = row.find_all('td')[2].get_text().strip()
        except:
            position = "NA"
        try:
            age = row.find_all('td')[3].get_text().strip()
        except:
            age = "NA"
        try:
            team_from = row.find_all('td')[5].get_text().strip()
        except:
            team_from = "NA"
        try:
            team_to = row.find_all('td')[6].get_text().strip()
        except:
            team_to = "NA"
        try:
            length = row.find_all("td")[7].get_text().strip()
        except:
            length = "NA"
        try:
            aav = row.find_all("td")[9].get_text().strip()
        except:
            aav = "NA"
        
        data.append((year, name, position, age, team_from, team_to, length, aav))
    
    year_count=year_count+1
    
    time.sleep(5)
    
df = pd.DataFrame(np.array(data))

df.columns = ['year','name','position','age','team_from','team_to','length','aav']

print(df)

df.to_csv('historical_contracts.csv')
```
